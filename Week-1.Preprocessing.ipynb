## 10. Handling Missing values

```python
for col in ["Temperature (C) - Min", "Temperature (C) - Max", 
            "Conductivity (Â¬Âµmho/cm) - Max","BOD (mg/L) - Max", 
            "NitrateN (mg/L) - Max", "Total Coliform (MPN/100ml) - Max"]:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())
    else:
        print(f"Column not found: {col}")
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
numeric_cols = df.select_dtypes(include='number').columns
df[numeric_cols] = imputer.fit_transform(df[numeric_cols])
```

```python
df.isnull().sum()
```

```python
# Remove extra spaces and non-ASCII characters
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace(r'[^\x00-\x7F]+', '', regex=True)
```
